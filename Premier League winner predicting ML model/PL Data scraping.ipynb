Scrapping data for multiple seasons and teams  with a loop

```python
#Loop to scrap data 

import requests
import pandas as pd
from bs4 import BeautifulSoup
from io import StringIO
import time
import random

# Initialize the list to store all match data
all_matches = []

# Initial URL for the 2023-2024 Premier League stats
initial_url = 'https://fbref.com/en/comps/9/2023-2024/2023-2024-Premier-League-Stats'

# Define the years you want to loop through
years = range(2024, 2022, -1)  # Example range, adjust as necessary

def fetch_page(url):
    try:
        response = requests.get(url)
        if response.status_code == 429:
            print("Rate limited. Sleeping for 60 seconds.")
            time.sleep(60)
            return fetch_page(url)  # Retry the request
        response.raise_for_status()
        return response
    except requests.RequestException as e:
        print(f"Request error: {e}")
        return None

for year in years:
    standings_url = initial_url  # Start with the initial URL
    retry_attempts = 0
    while retry_attempts < 5:
        try:
            # Fetch and parse the standings page
            response = fetch_page(standings_url)
            if not response:
                retry_attempts += 1
                continue

            soup = BeautifulSoup(response.text, 'html.parser')

            # Find the standings table
            standings_table = soup.select_one('table.stats_table')
            if not standings_table:
                print(f"No standings table found for year {year}.")
                break

            # Extract links to team pages
            links = [l.get("href") for l in standings_table.find_all('a') if l.get("href")]
            links = [l for l in links if '/squads/' in l]
            team_urls = [f"https://fbref.com{l}" for l in links]

            # Find the previous season link
            previous_seasons = soup.select("a.prev")
            if not previous_seasons:
                print(f"No previous season link found for year {year}.")
                break
            previous_season = previous_seasons[0].get("href")
            standings_url = f"https://fbref.com{previous_season}"

            # Process each team
            for team_url in team_urls:
                team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")

                # Fetch and parse the team page for match data
                response = fetch_page(team_url)
                if not response:
                    retry_attempts += 1
                    break

                html_content = response.text
                html_io = StringIO(html_content)
                matches = pd.read_html(html_io, match="Scores & Fixtures", flavor='lxml')[0]

                # Fetch and parse the shooting data
                soup = BeautifulSoup(response.text, 'html.parser')
                shooting_links = [l.get("href") for l in soup.find_all('a') if l.get("href")]
                shooting_links = [l for l in shooting_links if 'all_comps/shooting/' in l]
                if not shooting_links:
                    print(f"No shooting link found for team {team_name}.")
                    continue
                shooting_url = f"https://fbref.com{shooting_links[0]}"
                response = fetch_page(shooting_url)
                if not response:
                    retry_attempts += 1
                    break
                
                html_content = response.text
                html_io = StringIO(html_content)
                shooting = pd.read_html(html_io, match="Shooting", flavor='lxml')[0]
                shooting.columns = shooting.columns.droplevel()

                # Merge the data
                try:
                    team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
                    team_data = team_data[team_data["Comp"] == "Premier League"]
                    team_data["Season"] = year
                    team_data["Team"] = team_name
                    all_matches.append(team_data)
                except ValueError:
                    print(f"Error merging data for team {team_name}.")
                    continue

                time.sleep(random.uniform(2, 4))  # Randomized delay

            break  # Exit the retry loop after successful data collection

        except Exception as e:
            print(f"An error occurred: {e}")
            retry_attempts += 1
            time.sleep(60)  # Wait before retrying

# Convert the list of DataFrames to a single DataFrame
if all_matches:
    final_df = pd.concat(all_matches, ignore_index=True)
    print(final_df.head())
else:
    print("No match data collected.")
```
